```python
import os
import numpy as np
import matplotlib.pyplot as plt
from cartopy import config
import cartopy.crs as ccrs
from netCDF4 import Dataset as netcdf_dataset
```

```python
fname = os.path.join( 'data1.nc')
ds = netcdf_dataset(fname)
lats = ds.variables['latitude'][:]
long = ds.variables['longitude'][:]
time = ds.variables['time'][0]
```

```python
temp = ds.variables['TMP_2m'][:]
```

```python
lon, lat = np.meshgrid(long,lat)
ax = plt.axes(projection=ccrs.PlateCarree())
plt.contourf(lon, lat, temp[1,],60, trasform = ccrs.PlateCarree())

ax.coastlines()

plt.show()
```
